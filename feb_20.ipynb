Q1. Explain GET and POST methods.


GET: The GET method is used to request data from a specified resource. It sends the data in the URL as parameters. GET requests are typically used for retrieving data and should not have any side effects on the server.

POST: The POST method is used to submit data to be processed to a specified resource. It sends the data in the request body and can be used to create or update resources on the server.

Q2. Why is request used in Flask?


In Flask, the request object is used to access incoming request data, whether it's from a form submission, query parameters, or JSON data. It allows you to access the data sent by the client and process it in your Flask views.

Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the user to a different route or URL. It's often used after processing a form submission or performing some action to redirect the user to a different page, like a thank you page or a dashboard.

Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are used to separate the HTML presentation from the Python code. They allow you to dynamically generate HTML content by embedding placeholders that will be filled with data when the template is rendered.

The render_template() function in Flask is used to render HTML templates. It takes the template filename as an argument and can also pass data from the Python code to the template.
