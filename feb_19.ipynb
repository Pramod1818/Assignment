Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a micro web framework for Python that is designed to be lightweight and easy to use. It provides tools, libraries, and patterns to build web applications quickly and efficiently. Flask follows the WSGI (Web Server Gateway Interface) standard and is well-suited for small to medium-sized web applications and prototypes.

Advantages of Flask Framework:
Built-in Development Server: Flask comes with a built-in development server for testing purposes.
Flexibility: You can choose components you need (e.g., database, authentication) as per your project's requirements.
Extensible: Flask can be extended using various third-party extensions for common tasks.
Lightweight: Flask is minimalistic and doesn't impose unnecessary dependencies, making it easy to learn and use.
Easy to Get Started: Its simple syntax and structure make it beginner-friendly.
Well-Documented: Flask has comprehensive and easy-to-understand documentation.
Good for Prototyping: Ideal for building prototypes and small-scale applications quickly.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.


```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return '<h1>Hello World!!<h1>'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
    
```

Q3. What is App routing in Flask? Why do we use app routes?


App routing in Flask refers to defining URL routes that map to specific functions or views. When a user visits a URL, the associated function is executed. Routes are created using the @app.route() decorator.

Routes are used to define the structure of the application and determine how different URLs should be handled. They allow the application to respond to specific URLs with specific content or actions.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
Attach the screenshot of the output in Jupyter Notebook.


```python
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return '<h1>Welcome to ABC Corporation<h1>'

@app.route('/')
def company_info():
    return '''
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999<br>
    '''

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

The url_for() function in Flask is used to generate URLs for functions or views. It helps maintain dynamic URLs even if the route patterns change.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def hello():
    return 'Hello'

@app.route('/user/<username>')
def profile(username):
    return f'Profile for {username}'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('hello'))        
        print(url_for('profile', username='Raman'))  
```
