Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1. Bar Plot:
Use: Compare data across different categories or groups.

2. Scatter Plot:
Use: Visualize relationships between two numerical variables.

3. Box Plot:
Use: Display the distribution of data and identify outliers.

4. Heatmap:
Use: Display a matrix of data with color-coded cells to show patterns and correlations.

5. Pair Plot:
Use: Show pairwise relationships in a dataset with scatter plots and histograms.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("Line Plot of Signal over Timepoint")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.


```python
titanic = sns.load_dataset("titanic")

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Pclass")

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Pclass")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.


```python
diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title("Histogram of Price by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend(title="Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


```python
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue='species')
plt.title("Pair Plot of Iris Dataset")
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flt = sns.load_dataset("flights")

flt_piv = flt.pivot_table(index='month', columns='year', values='passengers')
plt.figure(figsize=(12, 6))
sns.heatmap(data=flt_piv, cmap='coolwarm', annot=True, linewidths=0.5)
plt.title("Heatmap of Flights Dataset")
plt.show()
```
