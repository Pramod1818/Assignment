Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have?


```python
import pandas as pd

# Load the dataset (replace 'flight_price_dataset.csv' with your file)
flight_data = pd.read_excel('data/flightprice.xlsx')

# Check dimensions
num_rows, num_columns = flight_data.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
flight_data.head() 
```

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.


```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(flight_data['Price'], bins=20, edgecolor='black')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?


```python
price_range = flight_data['Price'].max() - flight_data['Price'].min()
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

print(f"Price range: {price_range}")
print(f"Minimum price: {min_price}")
print(f"Maximum price: {max_price}")
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.


```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=flight_data, x='Airline', y='Price')
plt.xticks(rotation=45)
plt.xlabel('Airline')
plt.ylabel('Flight Price')
plt.title('Flight Price Variation by Airline')
plt.tight_layout()
plt.show()
```

Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
they may impact your analysis.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=flight_data, y='Price')
plt.ylabel('Flight Price')
plt.title('Boxplot of Flight Prices')
plt.show()
```

In the boxplot, outliers are typically points that are located far away from the central "box" of the plot. These points are represented as individual dots outside the "whiskers" of the boxplot.

Potential impacts of outliers on your analysis:

Skewed Averages: Outliers can significantly skew the mean and median values. If you're calculating average flight prices, outliers can lead to an overestimation or underestimation of the typical price.

Distorted Distributions: Outliers can distort the distribution of flight prices, making it difficult to understand the true spread of prices among the majority of flights.

Model Performance: If you're building predictive models, outliers can lead to model instability and decreased prediction accuracy. Models might give undue importance to outlier observations.

Decision-Making: Outliers could lead to incorrect decisions based on analysis. For example, if you're identifying pricing strategies, outliers might lead to misinterpretation of market trends.

Insights Misrepresentation: Outliers might indicate unique events that don't represent the usual behavior. Ignoring them or treating them as typical data points can lead to misinterpretation of insights.
